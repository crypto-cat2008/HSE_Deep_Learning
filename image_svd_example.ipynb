```python
from sklearn.decomposition import TruncatedSVD
import numpy as np
```

```python
for c in [1, 2, 10, 30]:
  svd = TruncatedSVD(n_components=c)
  cv2_imshow(np.clip(svd.inverse_transform(svd.fit_transform(img[:, :, 0])), 0, 255).astype(np.uint8))
```
